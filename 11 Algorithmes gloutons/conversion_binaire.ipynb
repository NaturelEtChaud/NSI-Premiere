## Conversion d'un nombre décimale en binaire

Le principe est, à chaque étape, de soustraire la plus grande puissance de 2 possible.

* Dans un premier temps on détermine la liste des puissances de 2 inférieures au nombre que l'on veut convertir

```python
def tab_puissances(nb):
    '''
    renvoie toutes les puissances de 2
    strictement inférieures à nb
    dans l'ordre décroissant
    '''
    tab = []
    i = 0
    while 2**i < nb :
        tab.append(2**i)
        i = i + 1
    tab.sort(reverse=True)
    return tab
```

* On peut maintenant procéder à la conversion par une méthode gloutonne

```python
def conversion_gloutonne(nb):
    '''
    donne l'écriture binaire d'un nombre
    par un algorithme glouton
    '''
    puissances = tab_puissances(nb)
    reste = nb
    i = 0
    binaire = ""
    while i<len(puissances):
        if reste - puissances[i] >= 0:
            reste = reste - puissances[i]
            binaire = binaire + "1"
            # cette ligne n'est pas nécessaire, elle permet de vérifier que l'on soustrait à chaque étape
            #la plus grande puissance de 2 possible
            print(puissances[i]) 
        else :
            binaire = binaire + "0"
        i = i + 1
    return binaire
```
