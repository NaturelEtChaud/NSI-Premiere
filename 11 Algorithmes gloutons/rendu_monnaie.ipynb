## Rendu de monnaie

Le principe est, à chaque étape, de soustraire la plus grande valeur possible, le plus de fois possible.

```python
def rendu(valeurs,somme):
    '''
    Entrées :
    valeurs est la liste des valeurs possibles dans l'ordre décroissant
    somme est la somme à rendre
    Sortie :
    les multiplicités de chaque valeur dont le résultat égale le nombre somme
    '''
    
    #nombre de types de pièces
    k = len(valeurs)
    #création du tableau des mutliplicités
    M = [0]*k
    
    reste = somme
    i = 0
    while reste > 0 and i < k :
        if reste - valeurs[i] >= 0 :
            # on peut rendre la ième valeur
            M[i] = M[i] + 1
            reste = reste - valeurs[i]
        else :
            #on passe à l'autre type de pièce
            i = i + 1
    return M
```

En utilisant le système monétaire `[5,2,1]`

```python
P = [5,2,1]
rendu(P,13)
```

Il faut donc 2 pièces de 5€, 1 pièce de 2€ et 1 pièce de 1€.

#### Malheureusement, l'algorithme glouton ne rend pas la solution optimale (avec le moins de pièce possible)

```python
P = [30, 24, 12, 6, 3, 1]
rendu(P, 49)
```

La solution proposée se compose de 4 pièces, alors que la solution optimale est `[0, 2, 0, 0, 0, 1]` avec uniquement 3 pièces.

#### Malheureusement, l'algorithme glouton peut se tromper est rendre une solution fausse (quand le résultat n'est pas possible)

```python
P = [30, 24, 12, 6, 3]
rendu(P, 49)
```

La solution proposée donne un total de 48€.

Une solution serait de vérifier que la réponse proposée est juste.

```python
def verificateur(valeurs, multiplicites, somme):
    '''
    Entrées
    valeurs est la liste des valeurs possibles dans l'ordre décroissant
    les multiplicites de chaque valeur
    (Ces deux tableaux sont de même dimension)
    somme est la somme désirée
    Sortie
    un booléen vrai si la somme des valeurs mutlipliées par leurs multiplicitées est égale à la somme
    '''
    calcul = 0
    for i in range(len(valeurs)):
        calcul = calcul + valeurs[i]*multiplicites[i]
    return calcul == somme
```

```python
def rendu_v2(valeurs,somme):
    '''
    Entrées :
    valeurs est la liste des valeurs possibles dans l'ordre décroissant
    somme est la somme à rendre
    Sortie :
    si le résultat est juste  : les multiplicités de chaque valeur dont le résultat égale le nombre somme
    sinon le nombre -1
    '''
    
    # recherche d'une solution
    multiplicites = rendu(valeurs, somme)
    
    # vérification
    if verificateur(valeurs, multiplicites, somme) :
        return multiplicites
    else :
        return -1
```

Deux essais :

```python
P = [5,2,1]
rendu_v2(P,13)
```

```python
P = [30, 24, 12, 6, 3]
rendu_v2(P, 49)
```
