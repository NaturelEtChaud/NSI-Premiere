1. Écrire une fonction `tableau_aleatoire()` qui renvoie un tableau de 20 valeurs prises aléatoirement entre 1 et 10.

```python
from random import randint

def tableau_aleatoire():
    tab = [0]*20
    for i in range(20):
        tab[i] = randint(1,10)
    return tab
```

```python
T = tableau_aleatoire()
#pour voir le contenu de T
T
```

2. Écrire une fonction `occurrence(tab,n)` qui prend en argument un tableau `tab` de valeurs prises entre 1 et 10 et l’argument `n` un nombre entre 1 et 10. Cette fonction devra compter le nombre d’occurrence du nombre `n` dans le tableau `tab`.

```python
def occurence(tab,n):
    compteur = 0
    for case in tab:
        if case==n:
            compteur = compteur + 1
    return compteur
```

```python
occurence(T,5)
```

3. Écrire une fonction `tableau_croissant(tab)` qui prend en argument un tableau `tab` de valeurs prises entre 1 et 10 et qui renvoie le tableau contenant tous les éléments du tableau `tab` sans aucune répétition, classé dans l’ordre. <br />
*(Comme nous n'avons pas le droit d'utiliser la méthode `append()` il faut trouver une solution qui n'est peut-être pas naturel au premier abord)*

```python
def tableau_croissant(tab):
    #le tableau present est composé de 11 cases booléennes
    #le nombre i est présent dans tab alors present[i] renverra True, sinon False
    
    #par défaut, le tableau present contient 11 False
    present = [False]*11
    #on regarde chaque nombre i entre 0 et 10.
    for i in range(11):
        #si ce nombre i est bien présent dans tab (occurence non nulle)
        if occurence(tab,i) != 0:
            #on annonce que ce nombre i est bien présent
            present[i] = True
    #on renvoie le résultat uniquement pour les nombres i présents
    return [i for i in range(11) if present[i]==True]
```

```python
tableau_croissant(T)
```

4. Écrire une fonction `jointure(tab1, tab2)` qui prend en arguments deux tableaux `tab1` et `tab2` de valeurs prises entre 1 et 10. Cette fonction retourne un tableau contenant tous les éléments de `tab1` et `tab2` sans aucune répétition, classé dans l’ordre. <br />
*(Il suffit d'effectuer la jointure de de `tab1` et `tab2` puis d'utiliser la fonction `tableau_croissant(tab)`)*

```python
def jointure(tab1,tab2):
    return tableau_croissant(tab1+tab2)
```
