**Exercice 1 :** Créer une fonction qui prend en argument un tuple de notes et qui renvoie un tuple donnant, dans cet ordre la plus petite note, la moyenne et la plus grande note.

Pour commencer, on va créer une tuple de notes.

```python
from random import randint

def notes():
    serie = ()
    nbNotes = randint(3,10)
    for i in range(nbNotes):
        nvlleNote = randint(0,20)
        serie = serie + (nvlleNote,)
    return serie
```

On va découper le problème en sous-problèmes plus simples. D'un côté une fonction pour calculer la moyenne, une fonction pour déterminer la note minimale et une dernière fonction pour déterminer la note maximale.

Pour la moyenne :

```python
def moyenne(serie):
    nbNotes = len(serie)
    #on prend la première note
    somme = serie[0]
    for i in range(1,nbNotes):
        somme = somme + serie[i]
    calcul = somme / nbNotes
    return calcul
```

Pour la note minimale :

```python
def minimum(serie):
    """
    avec une boucle pour
    """
    nbNotes = len(serie)
    candidat = serie[0]
    for i in range(1, nbNotes):
        if candidat > serie[i]:
            candidat = serie[i]
    return candidat
```

On peut néanmoins remarque qu'avec une boucle *POUR* on regarde toutes les notes, alors que si une note vaut 0 on a trouvé notre note minimale et ce n'est pas la peine de continuer la recherche.

```python
def minimum_v2(serie):
    """
    avec boucle tant que
    """
    nbNotes = len(serie)
    candidat = serie[0]
    i = 1
    while candidat != 0 and i<nbNotes:
        if candidat > serie[i]:
            candidat = serie[i]
        i = i + 1
    return candidat
```

Pour la note maximale :

```python
def maximum(serie):
    """
    avec boucle pour
    """
    nbNotes = len(serie)
    candidat = serie[0]
    for i in range(1, nbNotes):
        if candidat < serie[i]:
            candidat = serie[i]
    return candidat
```

De même, si on a une note égale à 20, on a trouvé notre maximum, ce n'est pas la peine de poursuivre la recherche.

```python
def maximum_v2(serie):
    """
    avec boucle tant que
    """
    nbNotes = len(serie)
    candidat = serie[0]
    i = 1
    while candidat != 20 and i<nbNotes:
        if candidat < serie[i]:
            candidat = serie[i]
        i = i + 1
    return candidat
```

Maintenant que tous les sous-problèmes sont reglés, on peut les regroupes dans une seule fonction.

```python
def statNotes(serie):
    m = minimum_v2(serie)
    moy = moyenne(serie)
    M = maximum_v2(serie)
    return m, moy, M
```

Testons notre travail.

```python
mes_notes = notes()

#pour voir le contenu de mes notes
print(mes_notes)

#on fait l'étude statistiques de mes notes
statNotes(mes_notes)
```
