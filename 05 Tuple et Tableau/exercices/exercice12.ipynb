**Exercice 12 :** Carré magique

Voir l'article wikipédia sur les carrés magiques : https://fr.wikipedia.org/wiki/Carr%C3%A9_magique_(math%C3%A9matiques)

```python
carre = [
    [4,9,2],
    [3,5,7],
    [8,1,6]
]
```

1. Écrire une fonction `somme_ligne(table,i)` qui calcule la somme de tous les nombres de la ligne `i`.

```python
def somme_ligne(table,i):
    ligne = table[i]
    somme = 0
    for case in ligne:
        somme = somme + case
    return somme
```

```python
somme_ligne(carre,2)
```

2. Écrire une fonction `somme_colonne(table,j)` qui calcule la somme de tous les nombres de la colonne `j`.

```python
def somme_colonne(table,j):
    somme = 0
    for ligne in table:
        somme = somme + ligne[j]
    return somme
```

```python
somme_colonne(carre,1)
```

3. Écrire une fonction `somme_diagonale_hgbd(table)` qui calcule la somme de tous les nombres de la diagonale du coin en haut à gauche au coin en bas à droite.

```python
def somme_diagonale_hgbd(table):
    somme = 0
    for i in range(len(table)):
        somme = somme + table[i][i]
    return somme
```

```python
somme_diagonale_hgbd(carre)
```

4. Écrire une fonction `somme_diagonale_bghd(table)` qui calcule la somme de tous les nombres de la diagonale du coin en bas à gauche au coin en haut à droite.

```python
def somme_diagonale_bghd(table):
    somme = 0
    n = len(table)
    for i in range(n):
        somme = somme + table[i][n-1-i]
    return somme
```

```python
somme_diagonale_bghd(carre)
```

5. Écrire un programme `carre_magique(table)` qui retourne `True` si la table est un carré magique.

```python
def carre_magique(table):
    magique = True
    #on teste les diagonales
    nb = somme_diagonale_bghd(table)
    if somme_diagonale_hgbd(table)!=nb:
        magique = False
    #si tout va bien, on continue en testant toutes les lignes
    i = 0
    while i<len(table) and magique == True :
        if somme_ligne(table,i)!=nb:
            magique = False
        i = i + 1
    #si tout va bien, on continue en testant toutes les colonnes
    j = 0
    while j<len(table) and magique == True :
        if somme_colonne(table,j)!=nb:
            magique = False
        j = j + 1
    return magique
```

```python
carre_magique(carre)
```
