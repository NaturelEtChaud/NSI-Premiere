**Exercice 6**

Écrire une fonction `cesar(chaine,clef)` prenant en argument chaine une chaîne de caractères uniquement entre a et z
(donc pas de majuscule, pas de chiffre, aucun caractère spéciaux, pas d'espace) et la clé `clef`, un nombre entier entre 1 et 25.

Chaque caractère de la chaîne est remplacé par celui obtenu après un décalage de `clef` caractères. <br />
Ainsi `cesar("bon",2)` doit renvoyer `"dqp"`. <br />
De même `cesar("xyz",2)` doit renvoyer `"zab"`. <br />

Pour de plus information sur le chiffre de César : https://fr.wikipedia.org/wiki/Chiffrement_par_d%C3%A9calage

```python
def cesar(chaine,clef) :
    '''
    chaine est une chaîne de caractères pris uniquement entre 'a' et 'z'
    '''
    crypt = ''
    for car in chaine :
        nouveau = ord(car)+clef
        #la lettre 'z' a pour code ASCII 122
        #pour gérer les dépassements, on décale de 26 autant que nécessaire
        while nouveau > 122 :
            nouveau = nouveau - 26
        crypt = crypt + chr(nouveau)
    return crypt
```

```python
cesar("bon",2)
```

```python
cesar("xyz",2)
```

**Remarque**

L'utilisation de la boucle **TANT QUE** semble plus intuitive, mais on aurait pu aussi utiliser le reste de la division entière. Le problème est que le caractrère 'a' n'est pas associé au nombre zéro.

```python
def cesar2(chaine,clef) :
    '''
    chaine est une chaîne de caractères pris uniquement entre 'a' et 'z'
    '''
    crypt = ''
    for car in chaine :
        nouveau = ord(car)+clef
        #la lettre 'a' a pour code ASCII 97
        #on soustrait ce nombre
        nouveau = nouveau - 97
        #on calcule le reste de la division entière par 26
        nouveau = nouveau % 26
        #on rajoute les 97
        nouveau = nouveau + 97
        
        #on aurait aussi pu directement faire
        #nouveau = ((nouveau - 97) % 26 ) + 97
        crypt = crypt + chr(nouveau)
    return crypt
```

```python
cesar2("bonxyz",2)
```
