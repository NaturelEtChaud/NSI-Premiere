Conversion d'un nombre décimale en nombre binaire

```python
def conversion(n):
    '''
    n est un nombre écrit en base 10
    le résultat est le même nombre écrit en binaire
    '''
    quotient = n
    resultat =""
    while quotient !=0:
        reste = quotient%2
        quotient = quotient//2
        resultat = str(reste) + resultat
    return resultat
```

```python
conversion(15)
```

```python
conversion(255)
```

```python
conversion(32)
```
