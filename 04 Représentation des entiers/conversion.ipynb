*Exercice 19 :* Conversion d'un nombre décimale en nombre binaire

```python
def conversion(n):
    '''
    n est un nombre écrit en base 10
    le résultat est le même nombre écrit en binaire
    '''
    quotient = n
    resultat =""
    while quotient !=0:
        reste = quotient%2
        quotient = quotient//2
        resultat = str(reste) + resultat
    return resultat
```

```python
conversion(15)
```

```python
conversion(255)
```

```python
conversion(32)
```

*Exercice 20 :* Conversion en base $b$

```python
def conversion(n,b):
    '''
    n est un nombre écrit en base 10
    le résultat est le même nombre écrit en base b
    '''
    quotient = n
    resultat =""
    while quotient !=0:
        reste = quotient%b
        quotient = quotient//b
        resultat = str(reste) + resultat
    return resultat
```

*Exercice 21 :* Conversion en base hexadécimale

```python
def conversion(n,b):
    '''
    Convertit n en base b avec une écriture spécifique pour la base 16
    '''
    quotient = n
    resultat =""
    while quotient !=0:
        reste = quotient%b
        quotient = quotient//b
        if reste == 10:
            reste = 'A'
        elif reste == 11:
            reste ='B'
        elif reste == 12:
            reste ='C'
        elif reste == 13:
            reste ='D'
        elif reste == 14:
            reste ='E'
        elif reste == 15:
            reste ='F'
        resultat = str(reste) + resultat
    return resultat
```

```python
conversion(255,4)
```

```python
conversion(666,16)
```

```python
conversion(6666,16)
```

*Exercice 22 :* Une autre méthode pour la conversion en base hexadécimale

```python
def conversion(n,b):
    '''
    Convertit n en base b avec une écriture spécifique pour la base 16
    '''
    hexa="ABCDEF"
    quotient = n
    resultat =""
    while quotient !=0:
        reste = quotient%b
        quotient = quotient//b
        #dans le cas d'une base hexadecimale
        if b==16 and reste >= 10:
            reste = hexa[reste-10]
        resultat = str(reste) + resultat
    return resultat
```

```python
conversion(255,4)
```

```python
conversion(666,16)
```

```python
conversion(6666,16)
```

*Exercice 23 :* Conversion en base Shadok

```python
def shadok(n):
    '''
    Convertit n dans la base Shadok, Ga Bu Zo Meu
    '''
    gabuzomeu=['Ga','Bu','Zo','Meu']
    quotient = n
    resultat =""
    while quotient !=0:
        reste = quotient%4
        quotient = quotient//4
        reste = gabuzomeu[reste]
        resultat = reste + resultat
    return resultat
```

```python
shadok(15)
```

```python
shadok(255)
```

```python
shadok(666)
```

*Exercice 24 :* Conversion en base 10

```python
def baseDecimale(n,b):
    '''
    Convertit le nombre n écrit en base b<=10 en base décimale
    '''
    nombre = str(n)
    longueur = len(nombre)
    resultat = 0
    for i in range(longueur-1,-1,-1):
        resultat = resultat + int(nombre[i])*b**(longueur-i-1)
    return resultat


#autre version
def baseDecimale2(n,b):
    nombre = str(n)
    longueur = len(nombre)
    resultat = 0
    for i in range(longueur-1):
        resultat = b*(resultat + int(nombre[i]))
    resultat = resultat + int(nombre[longueur-1])
    return resultat
```

```python
baseDecimale(10110,2)
```

```python
baseDecimale(1210,3)
```

```python
baseDecimale(666,10)
```

```python
baseDecimale(443,5)
```
