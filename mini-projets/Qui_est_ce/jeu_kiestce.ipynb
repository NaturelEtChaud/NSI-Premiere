## Question 4

Trouver tous les bruns qui ont une moustache ou une barbe

```python
for perso in kiestce :
    if perso['col_cheveux']=='brun' and (perso['barbe']=='moustache' or perso['barbe']=='barbe'):
        print(perso['nom'])
```

## Question 5
Ecrire un programme qui construit un tableau contenant tous les noms du jeu Qui est-ce ? <br />
Ecrire une fonction qui construit un tableau contenant tous les noms des personnages du jeu selon un critère.

```python
# pour récupérer tous les noms du jeu
def tab_noms(kiestce):
    T = ['']*(len(kiestce))
    for i in range (len (kiestce)) :
        T[i] = kiestce[i]['nom']
    return T
```

```python
tab_noms(kiestce)
```

```python
# une fonction qui renvoie un tableau contenant tous les noms des personnages selon un critère choisi
def tableau(kiestce, clef, valeur):
    taille=0
    for dico in kiestce :
        if dico[clef]==valeur:
            taille=taille+1
    T=['']*taille
    index = 0
    for dico in kiestce :
        if dico[clef]==valeur:
            T[index] = dico['nom']
            index = index + 1
    return T 
```

```python
# Pour avoir la liste des noms des filles
tableau(kiestce, 'genre', 'fille')
```
