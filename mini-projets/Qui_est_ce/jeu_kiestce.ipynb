 # Mini-projet - Qui est-ce ?

On dispose d'un tableau de dictionnaires contenant tous les personnages du célèbre jeu *Qui est-ce ?*

```python
from kiestce import kiestce
```

Pour chaque personnage, on a un dictionnaire aevc pour chaque clef suivante, on a l'une de ces valeurs :
* 'nom' : le nom du personnage
* 'genre' :  'garçon', 'fille'
* 'yeux' : 'marrons', 'bleus'
* 'cheveux' : 'courts', 'longs',  'chauve'
* 'coul_cheveux' : 'blanc','brun','roux','blond','chatain'
* 'nez' : 'petit','gros'
* 'barbe' : 'aucune', 'moustache', 'barbe'
* 'lunettes' : 'non', 'oui'
* 'chapeau' : 'non', 'oui'


## Question 1

Trouver tous les noms des personnages du jeu

```python
for perso in kiestce :
    print(perso['nom'])
```

## Question 2

Trouver le nom de toutes les filles

```python
for perso in kiestce :
    if perso['genre']=='fille':
        print(perso['nom'])
```

## Question 3

Trouver le nom de toutes les personnes aux yeix bleux et aux cheveux courts

```python
for perso in kiestce :
    if perso['yeux']=='bleus' and perso['cheveux']=='courts':
        print(perso['nom'])
```

## Question 4

Trouver tous les bruns qui ont une moustache ou une barbe

```python
for perso in kiestce :
    if perso['col_cheveux']=='brun' and (perso['barbe']=='moustache' or perso['barbe']=='barbe'):
        print(perso['nom'])
```

On pouait aussi faire la négation de ('moustache' ou 'barbe') qui correspond à 'aucune'

```python
for perso in kiestce :
    if perso['col_cheveux']=='brun' and not perso['barbe']=='aucune':
        print(perso['nom'])
```
