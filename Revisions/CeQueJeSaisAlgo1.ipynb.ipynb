# Tous les algorithmes de bases que je dois connaître

1) Je sais si c'est un entier est un nombre pair ou impair

(pas explicitement au programme mais très souvent utilisé)

```python
def est_pair(nb):
    if nb%2 == 0 :
        return True
    else :
        return False
```

2) Je sais rechercher le minimum d'un tableau d'entiers

```python
def minimum(tab):
    candidat = tab[0]
    for i in range(1,len(tab)):
        if candidat > tab[i]:
            candidat = tab[i]
    return candidat
```

```python
T = [1,5,8,6,21,3,4,7,9,989,521,0,2,64]
minimum(T)
```

3) Je sais faire la somme d'un tableau d'entiers

```python
def somme(tab):
    calcul = 0
    for i in range(len(tab)):
        calcul = calcul + tab[i]
    return calcul
```

```python
somme(T)
```

4) Je sais rechercher la posistion d'un élément dans un tableau d'entiers

(si le nombre n'est pas dans le tableau, on renvoie la valeur -1)

```python
def recherche(tab,nb):
    position = -1
    i = 0
    while (i<len(tab)) and (position == -1) :
        if tab[i] == nb :
            position = i
        i = i + 1
    return position
```

```python
recherche(T,989)
```

```python
recherche(T,988)
```
