```python
#les caractères et leur code
ord('a')               #renvoie le nombre décimal associé au caractère 'a'
chr(97)                #renvoie le caractère associé au nombre décimal 97
```

```python
#chercher dans une chaîne de caractères avec un indice
texte = 'bonjour les amibes !'
for i in range(len(texte)):
    if texte[i] == 'o':
        print('il y a un "o" dans ce texte.')
```

```python
#même chose mais en parcourant les caractères du texte (sans passage par indice)
texte = 'bonjour les amibes !'
for lettre in texte:
    if lettre == 'o':
        print('il y a un "o" dans ce texte.')
```
