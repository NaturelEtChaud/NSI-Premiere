## 1) Je sais les tableaux

```python
#construction d'un tableau des 10 entiers de 1 à 10
T = [0]*10
for i in range(len(T)):
    T[i] = i+1
```

```python
#même chose par compréhension
T = [i+1 for i in range(10)]
```

```python
#lecture d'un tableau avec l'indice, on affiche uniquement les nombres pairs
for i in range(len(T)):
    if T[i]%2 == 0:
        print(T[i])
```

```python
#même chose en parcourant les éléments du tableau (sans passage par indice)
for element in T:
    if element%2 == 0:
        print(element)
```

## 2) Je sais les dictionnaires

```python
#construction d'un dictionnaire à la main, chauqe couple clef/valeur est séparée par une virgule
adaDico =  {"prénom" : "Ada", "nom" : "Lovelace", "jour" : 10, "mois" : 12, "année" : 1815}
```

```python
#accès à une valeur, on met la clef entre crochets
adaDico['prénom']
```

```python
#on peut rajouter des couples en mettant la clef entre crochets et la valeur après le signe =
adaDico['titre'] = 'comtesse'
```

```python
#on peut construire un dictionnaire avec une boucle POUR
dico = {}
#ici la clef sera un entier, la valeur son carré
for i in range(10) :
    dico[i] = i**2
```

```python
#on peut construire un dictionnaire par compréhension
dico = {i:i**2 for i in range(10)}
```

```python
#on peut récupérer toutes les clefs avec une boucle POUR et la méthode keys()
for clef in dico.keys():
    print("J'ai la clef", clef)
```

```python
#on peut récupérer toutes les valeurs avec une boucle POUR et la méthode values()
for valeur in dico.values():
    print("J'ai la valeur", valeur)
```

```python
#on peut récupérer toutes lescouples clef/valeur avec une boucle POUR et la méthode items()
for clef, valeur in dico.items():
    print("Le carré de", clef, "est", valeur)
```

## 3)  On peut créer des structues plus complexes en combinant les structures de bases

Un dictionnaire de dictionnaires

```python
#les clefs 'naissance' et 'mort' renvoient vers un dictionnaire
adaDico = {"prenom" : "Ada",
    "nom" : "Lovelace",
    "naissance" : { "jour" : 10,     "mois" : 12,     "année" : 1815},
    "mort" : { "jour" : 27,     "mois" : 10,     "année" : 1852}
}
```

```python
#pour avoir accès à l'ensemble des informations concernant la naissance d'Ada
adaDico["naissance"]
```

```python
#pour avoir accès à l'année de naissance d'Ada
adaDico["naissance"]["année"]
```

Un tableau de dictionnaires

```python
stars = [
        {"prénom" : "Ada", "nom" : "Lovelace", "jour" : 10, "mois" : 12, "année" : 1815},
        {"prénom" : "Alan", "nom" : "Turing", "jour" : 23, "mois" : 6, "année" : 1912},
        {"prénom" : "Leonoardo", "nom" : "Fibonnaci", "jour" : None, "mois" : None, "année" : 1175},
        {"prénom" : "Aurélie", "nom" : "Jean", "jour" : 5, "mois" : 9, "année" : 1982}
        ]
```

```python
#pour avoir accès à l'ensemble des informations concernant Alan Turing, c'est le deuxième (donc i=1) élément du tableau
stars[1]
```

```python
#pour avoir accès à l'année de naissance d'Alan
stars[1]["année"]
```

```python
#et si on demandait le nom et prénom de celui ou celle qui est né en premier ????

#commentaires :
#(cela revient à rechercher le minimum dans un tableau, avec 2 ou 3 trucs en plus)
#ici on va considérer que chque personne est née une année différente
candidat = stars[0]["année"]
nom, prenom = stars[0]["nom"], stars[0]["prénom"]
for i in range(1,len(stars)):
    if stars[i]["année"] < candidat :
        candidat = stars[i]["année"]
        nom, prenom = stars[i]["nom"], stars[i]["prénom"]
print(prenom, nom, "est la personne qui est née en premier, en ",candidat)
```
