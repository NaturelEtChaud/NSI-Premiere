Écrire une fonction `print_ASCII(chaine)` qui affiche les codes ASCII au format hexadécimal d’une chaîne de caractère chaine.

```python
def print_ASCII(chaine):
    """
    les codes ASCII sont stockés
    dans un tableau
    """
    longu = len(chaine)
    T =[0]*longu
    for i in range(longu):
        T[i] = hex(ord(chaine[i]))
    return T
```

```python
print_ASCII("Ceci est mon texte.\nVoici une autre ligne.")
```
