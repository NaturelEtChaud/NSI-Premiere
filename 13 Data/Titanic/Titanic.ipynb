Ce fichier `titanic.csv` contient les descripteurs suivants :
* `PassengerId`, un numéro d’identifiant des passager ;
* `Survived`, 0 pour Non, 1 pour Oui ;
* `Pclass`, billet 1 re , 2 e ou 3 e classe ;
* `Name`, Nom et Prénoms du passager ;
* `Sex`, le genre, male ou female ;
* `Age`, l’âge en années ;
* `Sibsp`, nombre de frères, sœurs, demi-frères, demi-soœurs, époux, épouse à bord ; (les maîtresses et fiancées sont ignorées)
* `Parch`, nombre de parents ou d’enfants à bord ; (bien que certains enfants voyageaient avec une nurse parch=0)
* `Ticket`, numéro du ticket ;
* `Fare`, le tarif ;
* `Cabin`, numéro de la cabine ;
* `Embarked`, port d’embarquement ; C = Cherbourg, Q = Queenstown, S = Southampton.

## Importation d'un fichier CSV

```python
# pour lire en UTF8, n'est pas toujours nécessaire
import codecs

# ouverture du fichier en mode lecture 'r' pour 'read'
fichier = codecs.open("titanic.csv", 'r', 'utf-8')

# on crée la table de données
titanic = []

# on lit le fichier pour transférer les données dans la table
for ligne in fichier:
    # on enlève le retour à la ligne \r et la nouvelle ligne \n
    ligne = ligne.replace('\r','')
    ligne = ligne.replace('\n','')
    # on récupère chaque données de la ligne séparées par une virgule sous la forme d'un tableau
    ligne = ligne.split(',')
    # on rajoute ce tableau à notre table
    titanic.append(ligne)

# fermeture du fichier
fichier.close()
```

## Réponses aux questions de l'exercice
1. A l’aide de différentes fonctions, déterminer le nombre :
* de survivants
* de décédés
* de 1re classe
* de 2e classe
* de 3e classe
* de femmes survivantes en 1re classe
* d’hommes décédés en 3e classe

Pour le nombre de survivants :

```python
def nb_survivants(titanic):
    nb = 0
    for passager in titanic :
        if passager[1] == '1':
            # c'est un survivant
            nb = nb + 1
    return nb
```

```python
nb_survivants(titanic)
```

Pour le nombre de personnes décédées, nous avons deux solutions :
* Première solution : soit on calcule directement le nombre de personnes mortes (comme pour le nombre de survivants)

```python
def nb_morts(titanic):
    nb = 0
    for passager in titanic :
        if passager[1] == '0':
            # c'est un mort
            nb = nb + 1
    return nb
```

```python
nb_morts(titanic)
```

* Deuxième solution : on fait la soustraction entre le nombre de personnes dans le titanic moins le nombre de personnes survivantes. **Mais attention !**, le tableau `titanic` contient une ligne de plus que le nombre de passagers car il y a la première ligne qui est la ligne des descripteurs.

```python
def nb_morts_v2(titanic):
    nb_passagers = len(titanic) - 1
    return nb_passagers - nb_survivants(titanic)
```

```python
nb_morts_v2(titanic)
```

Pour le nombre de passagers par classe, on peut écrire une fonction qui renverra un tuple contenant les 3 valeurs attendues `(classe1, classe2, classe3)`.

```python
def classe(titanic):
    classe1, classe2, classe3 = 0, 0, -1
    for passager in titanic :
        if passager[2] == '1':
            # c'est un passger de 1ère classe
            classe1 = classe1 + 1
        elif passager[2] == '2':
            # c'est un passger de 2ième classe
            classe2 = classe2 + 1
        else :
            # c'est un passager de 3ième classe
            classe3 = classe3 + 1
    return classe1, classe2, classe3
```

```python
classe(titanic)
```

Pour les deux dernières dernières questions, on peut remarquer qu'elles se ressemble beaucoup. <br />
On va donc écrire une fonction qui aura pour arguments `sex`, `survived` et `pclass`.

```python
def requete(sex, survived, pclass):
    nb = 0
    for passager in titanic :
        if passager[5] == sex and passager[1] == survived and passager[2] == pclass:
            nb = nb + 1
    return nb
```

Ainsi pour le nombre de passagers femmes survivantes de 1ère classe

```python
requete('female', '1', '1')
```
