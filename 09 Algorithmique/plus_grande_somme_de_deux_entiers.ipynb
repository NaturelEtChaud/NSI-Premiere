 # Chapitre 9 - Algorithmique

Comparaison de deux méthodes pour effectuer la plus grande somme de deux nombres d'une liste. 

```python
from random import randint
#Ce sera notre liste de nombres pour ce travail
T = [randint(-50,50) for i in range(100)]
#POur voir ce qu'il y a dans la liste
T
```

## Première solution

On effectue toutes les sommes possibles et on garde la plus grande

```python
def somme2(tab):
    taille = len(tab)
    sum = tab[0] + tab[1]
    for i in range(taille):
        for j in range(taille):
            if i != j :
                if sum<tab[i]+tab[j]:
                    sum = tab[i]+tab[j]
    return sum
    
somme2(T)
```

## Deuxième solution

On cherche les deux plus grandes valeurs puis on les additionne

```python
def somme2_v2(tab):
    #une liste contenant le plus grand puis le deuxième plus grand
    if tab[0]>tab[1]:
        candidats = [tab[0], tab[1]]
    else :
        candidats = [tab[1], tab[0]]
    taille = len(tab)
    for i in range(2,taille):
        if tab[i] > candidats[1]:
            if tab[i] > candidats[0]:
                candidats[1], candidats[0] = candidats[0], tab[i]
            else :
                candidats[1] = tab[i]
    return candidats[0] + candidats[1]
    
somme2_v2(T)
```
