## 10 Green Bottles

```python
def green_bottles(n):
    '''
    n est le nombre de bouteilles au départ de la chanson
    '''
    chanson = ""
    for i in range(n,0,-1):
        chanson = chanson + 2*(str(i) + " green bottles hanging on the wall,\n")
        chanson =chanson + "And if one green bottle should accidentally fall,\n"
        if i > 1 :
            chanson = chanson + "There’ll be " + str(i-1) + " green bottles hanging on the wall.\n"
        else :
            chanson = chanson + "There’ll be no green bottles hanging on the wall."
    return chanson

#pour 10 bouteilles
paroles = green_bottles(10)
print(paroles)
```

## Alouette, gentil alouette

```python
#on peut compléter ce tableau...
T = ['la tête', 'le bec', 'le cou', 'le dos']

def alouette(n):
    assert n<=len(T), "n doit être inférieur à la taille du tableau qui fait ici "+str(len(T))
    chanson = ""
    for i in range(n):
        chanson = chanson + "Alouette, gentille alouette, alouette je te plumerai,\n"
        chanson =chanson + "Je te plumerai " + T[i] + ", je te plumerai " + T[i] + ",\n"
        for j in range(i,-1,-1):
            chanson = chanson + 2*("et " + T[j] +", ")
        chanson = chanson + "alouette, haaaa\n"
    return chanson

#pour 4 parties du corps de l'alouette
paroles = alouette(4)
print(paroles)
```
